```python
from network_full import *
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
test2=cv2.imread('test2.jpg',cv2.IMREAD_GRAYSCALE)
test1=cv2.imread('test1.jpg',cv2.IMREAD_GRAYSCALE)
plt.figure(1)
plt.imshow(test2,cmap='gray')
plt.figure(2)
plt.imshow(test1,cmap='gray')
```

```python

test2 = 255-cv2.resize(test2,(28,28))
test1 = 255-cv2.resize(test1,(28,28))
plt.figure(1)
plt.imshow(test2,cmap='gray')
plt.figure(2)
plt.imshow(test1,cmap='gray')
```

```python
model = Model.load('hand_written_digits.model')
```

```python
def data_treatment(image):
    return (image.reshape(image.shape[0]*image.shape[1]).astype(np.float32)-127.5)/127.5
```

```python
# pretreating the data
test2= data_treatment(test2)
test1=data_treatment(test1)
test2.shape
```

```python
confidences=model.predict(np.array([test2,test1]))
prediction=model.output_layer_activation.predictions(confidences)
prediction
```
